```python
import pandas as pd
path = "c:\\Users\\ADMIN\\Downloads\\introduction_pandas"
data = pd.read_csv(path + "/covid-19.csv")
data.head()
```

```python
data = data.iloc[:, 1:]
data.head()
```

```python
mask = data["geoId"] == "US"
us_data = data[mask]
us_data.head()
```

```python
us_data = us_data.reset_index(drop=True)
us_data.head()
```

```python
total_rows = len(us_data)
print(total_rows)
```

```python
us_data["cumulative_cases"] = 0
us_data.head()
```

```python
for i in range(total_rows-2, -1, -1):
    us_data.loc[i, "cumulative_cases"] = us_data.loc[i, "cases"] + us_data.loc[i + 1, "cumulative_cases"]
```

```python
us_data.head()
```

```python
us_data["cases_per_day"] = 0
us_data.head()
```

```python
for i in range(0, total_rows-1, 1):
    us_data.loc[i, "cases_per_day"] = us_data.loc[i, "cumulative_cases"] - us_data.loc[i+1, "cumulative_cases"]
```

```python
us_data.head()
```

```python
us_data["growth_ratio"] = 0
us_data.head()
```

```python
for i in range(0, total_rows-1, 1):
    us_data.loc[i, "growth_ratio"] = us_data.loc[i, "cases_per_day"] / us_data.loc[i+1, "cases_per_day"]
us_data.head()
```
